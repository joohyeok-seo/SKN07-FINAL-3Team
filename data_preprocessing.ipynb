# 산업체급식 Data

```python
food = df[df['식품기원명'] == '산업체급식(재료량 기반 산출 함량)']
```

```python
food.head()
```

```python
food['식품명'] = food['대표식품명']
```

```python
# 불필요한 col 제거
food.drop(columns=['식품기원명', '대표식품명', '데이터구분명', '식품중분류명', '식품소분류명'], inplace=True)
```

```python
# 식품명이 중복되는 값 확인
food[food.식품명.duplicated(keep=False)]
```

```python
# 식품중량 숫자만 추출
food['식품중량'] = food['식품중량'].str.extract(r'(\d+)', expand=False)
food['식품중량'] = food['식품중량'].astype(int)
```

```python
# 식품명이 중복 값 처리 > 평균값으로 대치
li = ['에너지(kcal)', '탄수화물(g)', '단백질(g)', '지방(g)', '당류(g)', '식이섬유(g)', '나트륨(mg)', '식품중량']

for i in li:
    food[i] = food.groupby('식품명')[i].transform('mean').round(2)
```

```python
# 평균값으로 대치 후 중복되는 행 제거
food = food.drop_duplicates()
```

```python
food.shape
```

```python
food.head()
```

```python
food.식품대분류명.value_counts()
```

```python
# 필요없는 식품 제거
food = food[~food['식품대분류명'].isin(['음료 및 차류', '장류, 양념류', '유제품류 및 빙과류', '곡류, 서류 제품'])]
```

```python
# 식품명이 국으로 끝나지만 다르게 분류되어 있는 경우 대분류 국으로 변경
food.loc[(food['식품명'].str.endswith('국')) & (food['식품대분류명'] != '국 및 탕류'), '식품대분류명'] = '국 및 탕류'
```

```python
food[food['식품대분류명'] == '면 및 만두류']
```

```python
# '식품명'에 '만두'가 포함된 경우만 필터링
food[food['식품명'].str.contains('만두')] 
```

```python
# 만두와 면 분리
food.loc[(food['식품명'].str.contains('만두')) & (food['식품대분류명'] == '면 및 만두류'), '식품대분류명'] = '만두류'
```

```python
# 만두를 제외한 나머지는 면류
food.loc[food['식품대분류명'] == '면 및 만두류'] = '면류'
```

```python
food.식품대분류명.value_counts()
```

```python
# 대분류명에 따른 식단 항목 매핑
category_map = {
    '밥': ['밥류'],
    '메인 반찬': ['볶음류', '조림류', '구이류', '찜류', '만두류'],
    '서브 반찬': ['튀김류', '전·적 및 부침류', '빵 및 과자류'],
    '서브 반찬2': ['나물·숙채류', '생채·무침류'],
    '국': ['국 및 탕류', '찌개 및 전골류', '죽 및 스프류', '면 및 만두류'],
    '김치': ['김치류', '장아찌·절임류', '젓갈류'],
}

def map_category(대분류명):
    for key, values in category_map.items():
        if 대분류명 in values:
            return key
    return '기타'  # 매칭되지 않는 경우 '기타'로 분류


food['카테고리'] = food['식품대분류명'].apply(map_category)
```

```python
food = food[['식품명', '대분류', '소분류', '기준량', '에너지(kcal)', '탄수화물(g)', '단백질(g)', '지방(g)', '당류(g)', '식이섬유(g)', '나트륨(mg)', '식품중량']]
```

```python
# 영양성분, 식품증량에 숫자가 아닌 데이터가 들어가 있는 row 삭제

tmp = ['탄수화물(g)', '단백질(g)', '지방(g)', '당류(g)', '식이섬유(g)', '나트륨(mg)', '식품중량']

food2 = food.copy()

for i in tmp:
    # 숫자인 값만 True로 반환 (정수 or 실수 문자열)
    mask = food2['식품중량'].astype(str).str.match(r'^[0-9]+(\.[0-9]+)?$')

    # 숫자가 값들만 필터링
    food2 = food2[mask]

    
food2
```
